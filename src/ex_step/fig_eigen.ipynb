```python
import h5py
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

```python
for h5 in glob.glob('./*h5'):
    h5file = h5py.File(h5,'r') ## Read list of outfiles
    x = h5file['x'][:]         ## x position
    y1 = h5file['y_1'][:]      ## y1 = y
    y2 = h5file['y_2'][:]      ## y2 = dy1/dx
    
    plt.title('Finite Potential Well')
    plt.xlabel('x')
    plt.plot(x, y1, label='y')
    plt.plot(x, y2, label='dy/dx')
    plt.legend(loc = 'best')
    plt.show()
    plt.close()
```
