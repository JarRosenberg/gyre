```python
import h5py
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

```python
for h5 in glob.glob('./*h5'):
    h5file = h5py.File(h5,'r')
    # print h5file.keys()
    x = h5file['x'][:]
    y1 = h5file['y_1'][:]
    y2 = h5file['y_2'][:]
    # print h5file['x'][:]
    # print h5file['y_1'][:]
    # print h5file['y_2'][:]

    plt.title('Particle in a Box')
    plt.xlabel('x')
    plt.plot(x, y1, label='y')
    plt.plot(x, y2, label='dy/dx')
    plt.legend(loc = 'best')
    plt.show()
    plt.close()
```
